## Fill in Area on Line Plots

In this exercise we work on a dataset that stores developers salaries by age. We will define an area on the plot and fill in that area. 

The `fill_between()` function generates a shaded region between a min and max boundary that is useful for illustrating ranges. We can use "where" argument to combine filling with logical ranges, e.g., to just fill in a curve over some threshold value.

```python
import matplotlib.pyplot as plt
import pandas as pd

salaries = pd.read_csv("Data/dev_salaries.csv")
salaries.head()
```

```python
ages = salaries["Age"]
dev_salaries = salaries["All_Devs"]
py_salaries= salaries["Python"]
js_salaries  = salaries["JavaScript"]

plt.plot(ages, dev_salaries, color="#444444", linestyle="--", label = "All Devs")

plt.plot(ages, py_salaries, label = "Python")

plt.legend()

plt.title("Median Salary (USD) by Age")
plt.xlabel("Ages")
plt.ylabel("Median Salary (USD)")

plt.tight_layout()
plt.show()
```

* Let's fill in the area below Python line.

`fill_between()` method: fills the area between two horizontal curves.
The curves are defined by the points (x, y1) and (x, y2). This creates one or multiple polygons describing the filled area. by default "y2" value equals to 0.  

```python
plt.plot(ages, dev_salaries, color="#444444", linestyle="--", label = "All Devs")

plt.plot(ages, py_salaries, label = "Python")

# "Alpha" defines the transparency. 
plt.fill_between(ages, py_salaries, alpha=0.3)

plt.legend()

plt.title("Median Salary (USD) by Age")
plt.xlabel("Ages")
plt.ylabel("Median Salary (USD)")

plt.tight_layout()
plt.show()
```

* Let's calculate the median of salaries of all developers and fill in the area between the line "Python" and overall median of salaries. We just need to pass in overall_meadian as "y2" argument in `fill_between()` method.

```python
import numpy as np

overall_median = np.median(dev_salaries)
print(overall_median)

plt.plot(ages, dev_salaries, color="#444444", linestyle="--", label = "All Devs")

plt.plot(ages, py_salaries, label = "Python")

# "Alpha" defines the transparency. 
plt.fill_between(ages, py_salaries, overall_median, alpha=0.3)

plt.legend()

plt.title("Median Salary (USD) by Age")
plt.xlabel("Ages")
plt.ylabel("Median Salary (USD)")

plt.tight_layout()
plt.show()
```

* We can add conditionals with "where" argument. Let's fill in the area where python salaries are greater than overall median. When we use "where" argument we need to set "interpolate" to True. Setting interpolate to True will calculate the actual intersection point and extend the filled region up to this point. 

```python
plt.plot(ages, dev_salaries, color="#444444", linestyle="--", label = "All Devs")

plt.plot(ages, py_salaries, label = "Python")

# "Alpha" defines the transparency. 
plt.fill_between(ages, py_salaries, overall_median,
                 where = (py_salaries > overall_median),
                 interpolate=True,
                 alpha=0.3)

plt.legend()

plt.title("Median Salary (USD) by Age")
plt.xlabel("Ages")
plt.ylabel("Median Salary (USD)")

plt.tight_layout()
plt.show()
```

* We can add additional conditionals.

```python
plt.plot(ages, dev_salaries, color="#444444", linestyle="--", label = "All Devs")

plt.plot(ages, py_salaries, label = "Python")

plt.fill_between(ages, py_salaries, overall_median,
                 where = (py_salaries > overall_median),
                 interpolate=True,
                 alpha=0.3)

plt.fill_between(ages, py_salaries, overall_median,
                 where = (py_salaries <= overall_median),
                 interpolate=True,
                 alpha=0.3)

plt.legend()

plt.title("Median Salary (USD) by Age")
plt.xlabel("Ages")
plt.ylabel("Median Salary (USD)")

plt.tight_layout()
plt.show()
```

* We can also fill in the area between two lines.(Phyton salaries and All developers' salaries.)

```python
plt.plot(ages, dev_salaries, color="#444444", linestyle="--", label = "All Devs")

plt.plot(ages, py_salaries, label = "Python")


# also add label
plt.fill_between(ages, py_salaries, dev_salaries,
                 where = (py_salaries > dev_salaries),
                 interpolate=True,
                 alpha=0.3,
                 label = "Above avg")

# also add label
plt.fill_between(ages, py_salaries, dev_salaries,
                 where = (py_salaries <= dev_salaries),
                 color='red',
                 interpolate=True,
                 alpha=0.3,
                 label = "Below avg")

plt.legend()

plt.title("Median Salary (USD) by Age")
plt.xlabel("Ages")
plt.ylabel("Median Salary (USD)")

plt.tight_layout()
plt.show()
```
