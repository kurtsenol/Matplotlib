## Subplots

In this exercise we will work on subplots. We use subplots to create more than one plots. Even, we can use it to create just one plot. 

```python
import matplotlib.pyplot as plt
import pandas as pd

devs = pd.read_csv("Data/dev_salaries.csv")
devs.head()
```

```python
ages = devs["Age"]
all_salaries = devs["All_Devs"]
py_salaries = devs["Python"]
js_salaries = devs["JavaScript"]
```

* We will use `subplots()`(note the "s" at the end) method to create a figure and certain number of axes. If we don't pass in rows and columns to the `subplots()` method, it will create just one axis.  

We use;
* plot() method of axes object instead of the plot() method of pyplot object: `ax.plot()` instead of `plt.plot()`

* `ax.legend()` instead of `plt.legend()`
* `ax.set_title()` instead of `plt.title()`
* `ax.set_xlabe()` instead of `plt.xlabel()`
* `ax.set_ylabe()` instead of `plt.ylabel()`
* No change for `plt.tight_layout()` and `plt.show()` methods

```python
fig, ax = plt.subplots()

ax.plot(ages, py_salaries, label = "Python")
ax.plot(ages, js_salaries, label = "JavaScript")
ax.plot(ages, all_salaries, label = "All_Devs",
       linestyle= "--", c= "#444444")

ax.legend()

ax.set_title("Median Salary (USD) by Age")
ax.set_xlabel("Ages")
ax.set_ylabel("Median Salary (USD)")

plt.tight_layout()
plt.show()
```

* Let's look at what we get from "ax" and "fig" when we run `plt.subplots()` method.

```python
fig, ax = plt.subplots()
print(ax)
print(fig)
```

* What if we pass in 2 rows and 3 columns:

```python
fig, ax = plt.subplots(nrows=2, ncols=3)
print(ax)
print(fig)
```

* As you can see "ax" variable equals to number of plots that we pass in as an argument (defined by number of rows and columns) to the `plt.subplots()` method. All plots are displayed on the same figure which is the overall window.      

* We can also unpack axes object instead of assigning all to a single "ax" variable. 

```python
fig, ((ax_1, ax_2, ax_3), (ax_4, ax_5, ax_6)) = plt.subplots(nrows=2, ncols=3)
print(ax_1)
print(fig)
```

* Let's display a plot by unpacking the axes.

```python
fig, ((ax_1, ax_2, ax_3), (ax_4, ax_5, ax_6)) = plt.subplots(nrows=2, ncols=3)

ax_1.plot(ages, all_salaries, label = "All_Devs",
       linestyle= "--", c= "#444444")

ax_1.legend()

ax_1.set_title("Median Salary (USD) by Age")
ax_1.set_xlabel("Ages")
ax_1.set_ylabel("Median Salary (USD)")

plt.tight_layout()
plt.show()
```

* or we can get axes back as a numrows x numcolumns object array which supports numpy indexing. Let's display a plot with this method.

```python
fig, ax = plt.subplots(nrows=2, ncols=3)

ax[0,0].plot(ages, all_salaries, label = "All_Devs",
       linestyle= "--", c= "#444444")

ax[0,0].legend()

ax[0,0].set_title("Median Salary (USD) by Age")
ax[0,0].set_xlabel("Ages")
ax[0,0].set_ylabel("Median Salary (USD)")

plt.tight_layout()
plt.show()
```

* Now, let's return to our dataset and break down our initial plot into two; one shows all developers' salaries and the other shows salaries of python and javascript developers together. in a different plot  

```python
fig, (ax_1, ax_2) = plt.subplots(nrows=2, ncols=1)

ax_1.plot(ages, all_salaries, label = "All_Devs",
       linestyle= "--", c= "#444444")

ax_2.plot(ages, py_salaries, label = "Python")
ax_2.plot(ages, js_salaries, label = "JavaScript")

ax_1.legend()

ax_1.set_title("Median Salary (USD) by Age")
ax_1.set_xlabel("Ages")
ax_1.set_ylabel("Median Salary (USD)")

ax_2.legend()

ax_2.set_title("Median Salary (USD) by Age")
ax_2.set_xlabel("Ages")
ax_2.set_ylabel("Median Salary (USD)")

plt.tight_layout()
plt.show()
```

If we look the figure above, we can make some adjustments;
* remove the title for ax_2,
* remove the x_label for ax_1,
* remove the X ticks in ax_1 by passing in "sharex = True" argument to the `plt.subplots()` method. We can also use "sharey = True" argument when we display more than one axes in a row.

* Let's apply these changes and check if it looks better!

```python
fig, (ax_1, ax_2) = plt.subplots(nrows=2, ncols=1, sharex=True)

ax_1.plot(ages, all_salaries, label = "All_Devs",
       linestyle= "--", c= "#444444")

ax_2.plot(ages, py_salaries, label = "Python")
ax_2.plot(ages, js_salaries, label = "JavaScript")

ax_1.legend()

ax_1.set_title("Median Salary (USD) by Age")
ax_1.set_ylabel("Median Salary (USD)")

ax_2.legend()

ax_2.set_xlabel("Ages")
ax_2.set_ylabel("Median Salary (USD)")

plt.tight_layout()
plt.show()
```
